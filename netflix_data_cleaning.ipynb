# Deleting redundant column

```python
data.drop(columns = "rating", inplace = True)
```

```python
data.columns
```

# Renaming columns

```python
# Create an empty list to store capitalized column names
new_column_names = []
# Iterate through each column name in the dataframe
for i in data.columns:
    # Capitalize the first letter of each column name and add it to the list
    new_column_names.append(i.capitalize())
```

```python
data.columns = new_column_names
```

```python
data
```

# Dropping duplicates

```python
# Check for duplicate rows in the dataframe
# Returns a boolean Series where True indicates the row is a duplicate of a previous row
data.duplicated()
```

```python
# Check for duplicated rows in the dataframe and count the total number of duplicates
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data
```

# Remove the NaN values from the dataset

```python
# Check for missing values in the dataframe
# Returns a boolean DataFrame where True indicates missing values (NaN)
data.isna()
```

```python
# Check the number of missing values in each column of the dataframe
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

# Some more transformation

```python
# Get all unique show IDs from the 'Show_id' column in the data DataFrame
# This helps identify how many distinct shows are in the dataset
data["Show_id"].unique()
```

```python
# The 'Show_id' column contains values like 's1234'.
# This line removes the leading 's' character and extracts only the numeric part.
# It splits each value at 's' and keeps the portion after it,
# converting identifiers into clean numeric-like strings for easier analysis.
data["Show_id"] = data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
type(data["Show_id"][0])
```

```python
# Convert the 'Show_id' column to integer data type
data["Show_id"].astype(int)
```

```python
# Convert the 'Show_id' column to integer data type and saving it to original dataframe
data["Show_id"] = data["Show_id"].astype(int)
```

```python
data
```

```python
# The 'Date_added' column contains dates formatted with forward slashes (e.g., '09/25/2020').
# This line standardizes the date format by replacing '/' with '-',
# ensuring consistent formatting across all rows before datetime conversion.
# A uniform date format helps avoid parsing errors during further processing.
data["Date_added"] = data["Date_added"].apply(lambda x: x.replace("/", "-"))
```

```python
data
```

```python
# Count the frequency of each unique country in the dataset and return as a Series sorted by count in descending order
data["Country"].value_counts()
```

```python
data.to_csv("cleaned_netflix.csv", index = False)
```
